```python
#Eliminar Duplicados
tweets_df = tweets_df.drop_duplicates()
```

```python
#Comprobamos si existen caracteres repetidos dentro de los tweets
tweets_df[tweets_df['text'].str.contains("(.)\\1{2,}")].head(3)
```

```python
#Comprobamos si existen palabras compuestas mediante guion repetidos dentro de los tweets
tweets_df[tweets_df['text'].str.contains("([A-Za-z]+)-([A-Za-z]+)")].head(3)
```

```python
#Comprobamos si existen numeros dentro de los tweets
tweets_df[tweets_df['text'].str.contains("r\d+")].head(8)
```

```python
#Implementamos una funcion para poder tratar estas casuisticas y transformarlas a texto simplificado

def preprosess(text):
    text = text.lower()
    text = re.sub(r'@\w+', '', text) #menciones
    text = re.sub('\n',' ',text) #saltos de linea
    text = re.sub('#[A-Za-z0-9]+\s?','',text) #hashtags
    text = re.sub('[^\w\s]','',text) #puntuacion
    text = re.sub('"(.)\\1{2,}"','\\1\\1',text) #caracteres repetidos
    text = re.sub("([A-Za-z]+)-([A-Za-z]+)", "\\1\\2", text) #palabras compuestas
    text = re.sub("^\d+\s|\s\d+\s|\s\d+$",' ', text)# numeros
    text = re.sub('url','',text) #saltos de linea
    text = re.sub("mas movil",'',text)
    text = re.sub("mÃ¡s movil",'',text)
    text = re.sub("mÃ¡s mÃ³vil",'',text)
    text = re.sub("mas mÃ³vil",'',text)
    text = re.sub("movistar",'',text)
    text = re.sub("telefÃ³nica",'',text) 
    text = re.sub("orange",'',text) 
    text = re.sub("yoigo",'',text) 
    text = re.sub("jazztel",'',text) 
    text = re.sub("vodafone",'',text)
    text = unidecode.unidecode(text)#para quitar tildes
    return text
```

```python
tweets_df['text'] = tweets_df['text'].apply(preprosess)
tweets_df.head(10)
```

## TokenizaciÃ³n

```python
#Lista de los comentarios de la columna Comentario
tweets = tweets_df['text'] .values.tolist()
tweets
```

```python
#Tokenizamos los comentarios obtenidos del dataset
import nltk
from nltk.tokenize import word_tokenize
nltk.download('punkt')

tweets_token = []
for tw in tweets:
    tweets_token.append(word_tokenize(tw))

tweets_token[0] #imprimo la primera linea
```

```python
tweets_token
```

## CorrecciÃ³n OrtogrÃ¡fica

```python
#!pip install symspellpy
```

```python
import pkg_resources
from symspellpy import SymSpell, Verbosity
# Busca el diccionario y lo configura
sym_spell = SymSpell(max_dictionary_edit_distance=2, prefix_length=7)
dictionary_path = pkg_resources.resource_filename("symspellpy", 
"es-100lANSI.txt")
# "term_index" es la columna del termino y "count_index" es la de la frecuencia
sym_spell.load_dictionary(dictionary_path, term_index=0, count_index=1)
# Busca la correccion de la palabra
# Si es correcta o no encuentra correccion devuelve la palabra original
correct_tokens = []
for m in tweets_token:
 correct_word = [sym_spell.lookup(w, Verbosity.TOP, include_unknown=True)[0].term 
for w in m]
 correct_tokens.append(correct_word)
```

## EliminaciÃ³n StopWords

```python
#Eliminamos las stopwords existentes
from nltk.corpus import stopwords
from unidecode import unidecode

nltk.download('stopwords')

stop = set(stopwords.words('spanish')) #Fijamos el idioma
# Inicializa la lista
correct_tokens = [[unidecode(palabra) for palabra in array] for array in correct_tokens]

tweets_stop = []
# Solo guarda en la lista las palabras que no son stopwords
for tw in correct_tokens:
 l = [c for c in tw if not c in stop]
 tweets_stop.append(l)


```

```python
tweets_stop
```

## ObtenciÃ³n RaÃ­z de las Palabras

```python
#Modulo PorterStemmer para obtener la raiz de las palabras
#TODO Revisar que funciona bien en base a los pasos anteriores
from nltk.stem import SnowballStemmer

 # Inicializa la lista
stem_tokens = [] 
 # Crea el objeto "SnowballStemmer"
stemmer = SnowballStemmer('spanish')
 # Guarda en la lista las raices de las palabras
for l in tweets_stop: #he cambuado tokens por stop
    b = [stemmer.stem(q) for q in l]
    stem_tokens.append(b)

```

```python
stem_tokens
```

### ActualizaciÃ³n Dataset

```python
tweetsStopString = []
for arrStop in tweets_stop:
    s = " ".join(arrStop)
    tweetsStopString.append(s)
    
tweetsStopString
```

```python
tweets_clean = []
for tw in stem_tokens:
    tweets_clean.append(" ".join(tw))
    
tweets_clean
```

```python
#reviews_df.assign(Preprocedado = comentarios_clean)
tweets_df["Texto Preprocesado"] = tweetsStopString
tweets_df["Raiz Palabras"] = tweets_clean
tweets_df
```
